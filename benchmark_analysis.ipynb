```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
threads_vs_size = pd.read_csv('benchmarking/threads_vs_size.csv', index_col=0)
threads_vs_size
```

```python
block_size_vs_size = pd.read_csv('benchmarking/block_size_vs_size.csv', index_col=0)
block_size_vs_size
```

```python
fig = plt.figure(figsize=(10, 7.5))
for thread in range(1, 11):
    plt.plot([int(col) for col in threads_vs_size.columns], threads_vs_size.loc[thread], label=f'{thread} threads')

plt.ylabel('Time (microseconds)', fontsize=15)
plt.xlabel('Length of one sequence', fontsize=15)
plt.legend(fontsize=13, loc='upper left')
plt.title('Threads vs length of one sequence', fontsize=18)
plt.savefig('benchmarking/threads_vs_size.png')
```

```python
fig = plt.figure(figsize=(10, 7.5))
for thread in range(1, 11):
    plt.plot([int(col)*int(col) for col in threads_vs_size.columns], threads_vs_size.loc[thread], label=f'{thread} threads')

plt.ylabel('Time (microseconds)', fontsize=15)
plt.xlabel('Size of matrix', fontsize=15)
plt.legend(fontsize=13, loc='upper left')
plt.title('Threads vs size of matrix', fontsize=18)
plt.savefig('benchmarking/threads_vs_size_matrix.png')
```

```python
fig = plt.figure(figsize=(10, 7.5))
for size in range(500, 2001, 500):
    plt.plot(block_size_vs_size.index, block_size_vs_size.loc[:, str(size)], label=f'Sequence length: {size}')

plt.ylabel('Time (microseconds)', fontsize=15)
plt.xlabel('Block size (length of one side)', fontsize=15)
plt.legend(fontsize=13, loc='upper right')
plt.title('Length of block vs length of sequence', fontsize=18)
plt.savefig('benchmarking/block_size_vs_size.png')
```
